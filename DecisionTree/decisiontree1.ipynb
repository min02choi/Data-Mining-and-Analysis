```python
# 데이터 불러오기
train = pd.read_csv('/Users/min02choi/IdeaProjects/Data-Mining-and-Analysis/train.csv')
test = pd.read_csv('/Users/min02choi/IdeaProjects/Data-Mining-and-Analysis/test.csv')
train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
#no_of_adults,no_of_children,avg_price_per_room 특징만 고려
feature_names = ["no_of_adults", "no_of_children", "avg_price_per_room"]
X = train[feature_names]
y = train["booking_status"]
```

```python
# Decision Tree 학습/예측/평가
X_train, X_test, y_train, y_test = train_test_split(X,y)
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
dt_pred = dt_model.predict(X_test)
print('DecisionTreeClassifier 정확도: {0:.4f}'.format(accuracy_score(y_test, dt_pred)))
```
